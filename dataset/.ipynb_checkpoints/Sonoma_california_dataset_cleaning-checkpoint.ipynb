```python
# Our data management libraries
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('../dataset/Sonoma_County_California_dataset.csv')
df.head()
```

```python
#dropping columns we don't need
columns_to_drop = ['Name', 'Impound Number','Kennel Number',
                   'Animal ID','Intake Subtype','Outcome Subtype','Outcome Condition',
                  'Intake Jurisdiction','Outcome Jurisdiction','Outcome Zip Code','Location','Count']
df.drop(columns=columns_to_drop, inplace=True)
df.head(50)
```

```python
#after getting rid of columns I don't need, I check for nulls
df.isnull().sum()
```

```python
#drop nulls
df = df.dropna()
```

```python
# I check for nulls again
df.isnull().sum()
```

```python
#convert DOB into age
# Convert 'DOB' column to datetime format
df['Date Of Birth'] = pd.to_datetime(df['Date Of Birth'])

# Calculate age based on the current date
current_date = datetime.now()
df['Age'] = (current_date - df['Date Of Birth']).astype('<m8[Y]').astype(int)
df.head(50)
```

```python
#Now that we calculated age with DOB, we drop it. We also don't need dates since we have Days in Shelter
columns_to_drop = ['Date Of Birth','Intake Date','Outcome Date']
df.drop(columns=columns_to_drop, inplace=True)
df.head(50)
```

```python
#It is clean, so we export
df.to_csv("../dataset/clean_Sonoma_Cali_dataset.csv", index = False)
```
