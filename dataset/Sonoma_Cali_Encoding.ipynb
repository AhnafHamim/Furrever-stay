```python
# Our data management libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../dataset/clean_Sonoma_Cali_dataset.csv')
df.head(50)
```

# Encoding Color

```python
# Define a function to encode colors
def encode_color(color):
    if 'TABBY' in color:
        return 'TABBY'
    elif color in ['BLACK', 'BROWN', 'ORANGE', 'WHITE', 'GRAY', 'CALICO']:
        return color
    else:
        return 'OTHER'

# Apply the function to create a new 'Encoded Color' column
df['Encoded Color'] = df['Color'].apply(encode_color)

# Use LabelEncoder to encode 'Encoded Color' column
label_encoder = LabelEncoder()
df['Encoded Color'] = label_encoder.fit_transform(df['Encoded Color'])
```

# ENCODING BREED

```python
# Define a function to encode colors
def encode_breed(breed):
    if 'DOMESTIC SH' in breed:
        return 'DOMESTIC SH'
    elif breed in ['DOMESTIC MH','DOMESTIC LH']:
        return breed
    else:
        return 'OTHER'

# Apply the function to create a new 'Encoded Color' column
df['Encoded Breed'] = df['Breed'].apply(encode_breed)

# Use LabelEncoder to encode 'Encoded Color' column
label_encoder = LabelEncoder()
df['Encoded Breed'] = label_encoder.fit_transform(df['Encoded Breed'])
```
