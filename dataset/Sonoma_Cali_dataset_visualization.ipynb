```python
# Our data management libraries
import pandas as pd
import numpy as np

# A basic visualization library
import matplotlib.pyplot as plt

# A great visualization library
import seaborn as sns

# Very important, this will make your charts appear in your notebook instead of in a new window.
%matplotlib inline
from datetime import datetime


# Seaborn / matplotlib for visualization 
import seaborn as sns
sns.set()

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../dataset/clean_Sonoma_Cali_dataset.csv')
df.head(50)
```

# Colors

```python
category_counts = df['Color'].value_counts()
category_counts.plot(kind='bar', color='yellow', edgecolor='black', figsize=(30, 25))

# Adding labels and title
plt.xlabel('Color')
plt.ylabel('Frequency')
plt.title('Color Data Distribution')

# Display the plot
plt.show()
```

# Days in Shelter x Intake Condition

```python
condition_means = df.groupby('Intake Condition')['Days in Shelter'].mean()

# Plot a bar chart
condition_means.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(15,8))

# Adding labels and title
plt.xlabel('Intake Condition')
plt.ylabel('Average Days in Shelter')
plt.title('Average Days in Shelter by Intake Condition')
```
