**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Loading the dataset**

```python
df = pd.read_csv('df_only_numeric_value_nov_05.csv')
df.head()
```

**Scatter plot for days_stayed vs Age**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['days_stayed'], alpha=0.5)  # alpha is set to see the density of points
plt.title('Scatter Plot of Days Stayed vs Age')
plt.xlabel('Age')
plt.ylabel('Days Stayed')
plt.show()
```

**Scatter plot for days_stayed vs Primary_Color**

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['Primary_Color_encoded'], df['days_stayed'], alpha=0.5)
plt.title('Scatter Plot of Days Stayed vs Primary Color Encoded')
plt.xlabel('Primary Color Encoded')
plt.ylabel('Days Stayed')
plt.show()
```
